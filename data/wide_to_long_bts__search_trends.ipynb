```python
# import data from csv file
import pandas as pd
data = pd.read_csv('bts_kpop_search_trend.csv')
data.head()
                   
```

```python
# convert data from wide to long format
data_long = pd.melt(data, id_vars = ['Week'], var_name = 'artist', value_name = 'search_popularity_trends')
data_long.head()
```

```python
# Convert the 'Week' column to a proper datetime format
data['Week'] = pd.to_datetime(data['Week'], format='%d/%m/%Y')

# Check the updated data
data.head()
```

```python
# export data to new csv file
data_long.to_csv('bts_kpop_search_trend_long.csv', index = False)
```
